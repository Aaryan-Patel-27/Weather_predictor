```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load your weather dataset (make sure to adjust the path)
file_path = r"C:\Users\Asus\Desktop\DATA_WEATHER.csv"
df = pd.read_csv(file_path)
```

```python
temp_summary = df.groupby("Daily Summary")["Temperature (C)"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
temp_summary.plot(kind="barh", color="skyblue")
plt.xlabel("Sum of Temperature (C)")
plt.title("Sum of Temperature (C) by Daily Summary")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Visibility (km)", y="Temperature (C)", hue="Daily Summary", data=df, palette="Set2", legend=False)
plt.title("Sum of Visibility (km) by Temperature (C)")
plt.xlabel("Sum of Visibility (km)")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
temp_summary.plot(marker='o')
plt.ylabel("Sum of Temperature (C)")
plt.title("Sum of Temperature (C) by Daily Summary")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Limit to top 10 for visibility
top_temp = temp_summary.head(10)

plt.figure(figsize=(8,8))
plt.pie(top_temp, labels=top_temp.index, autopct='%1.1f%%', startangle=140)
plt.title("Sum of Temperature (C) by Daily Summary")
plt.tight_layout()
plt.show()
```
